Key Metrics

We embarked on a comprehensive analysis of key metrics across platforms, focusing on engagement rates, reach, and user demographics.

•	Engagement Rates
Our analysis of engagement rates unveiled noteworthy insights:
Instagram: Instagram consistently displayed the highest engagement rate, peaking at 10%. Users on this platform appear to be highly engaged with the content.

Facebook: Facebook closely followed Instagram in engagement rate, averaging around 9%.

LinkedIn and Twitter: These platforms exhibited lower engagement rates, hovering at approximately 5% and 4%, respectively. Users on these platforms may require different engagement strategies.

Reach
Our reach analysis revealed the following:
Instagram: Instagram boasted the highest reach, indicating a wide user base. It consistently reached more users than other platforms.
Facebook: Facebook, while having slightly lower reach than Instagram, still reached a substantial audience.
Twitter and LinkedIn: These platforms had comparatively lower reach, suggesting the need for targeted approaches.

Engagement Analysis
In-depth analysis of engagement metrics, including 'Engagements,' 'Reactions,' 'Likes,' 'Comments,' 'Shares,' and 'Post Clicks (All),' provided insightful findings:

'Likes' and 'Reactions': Across all platforms, 'Likes' and 'Reactions' were the most common forms of engagement. They constituted a significant portion of the engagement chart.

Facebook: Facebook stood out with the highest 'Comments' and 'Shares,' indicating that users on this platform are more inclined to interact and share content, followed by Twitter.

Instagram: Instagram excelled in 'Likes,' suggesting that its user base favors quick interactions through 'Likes.'

Post Types
Categorizing posts into types yielded intriguing results:
'Image' and 'Video' Posts: 'Image' and 'Video' posts consistently garnered the highest engagement across all platforms, indicating that visual content is more appealing to users.

The Images were mostly posted on Facebook, Video post were more shared on Twitter, same as text with more engagement on Twitter.
'Status' Posts: 'Status' posts had the lowest engagement rates. This suggests that users may find longer text-based posts less engaging.

Post Clicks (All)There are more of post click on Linkedln with content like polls and documents, they shows the kind of content with better engagement on linkedln.

Key Findings
Our analysis unveiled several critical findings:

Most Engaging Types of Posts
'Image' and 'Video' posts consistently garnered the highest engagement across all platforms. These content formats resonate well with users and should be prioritized in content strategies.

'Status' posts had the lowest engagement rates. Playhouse Communication may consider reevaluating the use of this post type.

Platform with Highest Engagement
Facebook emerged as the platform with the highest overall engagement rate such as Comments, Shares, Post Clicks (All) showcasing the potential for impactful marketing campaigns.

Instagram is the platform with some much reaction such as Likes.
Peak Times for User Engagement

User engagement peaks during late evenings and some early hour of the day. Posting content during these hours can maximize engagement. This insight is critical for content scheduling.

Trends Over Time
We observed an upward trend in 'Likes' across all platforms over the ten-year period, indicating that users increasingly favor this form of engagement.
'Shares' showed a declining trend, suggesting a shift in user behavior, potentially toward more passive consumption.

```python
# Define the engagement metrics
engagement_metrics = ['Engagements', 'Reactions', 'Likes', 'Comments', 'Shares', 'Post Clicks (All)']


# Plot histograms for engagement metrics
for metric in engagement_metrics:
    # Check for duplicate values in the specified metric column
    if combined_data[metric].duplicated().any():
        # If duplicates exist, consider aggregating the data, e.g., by summing it for each unique value
        aggregated_data = combined_data.groupby(metric).sum().reset_index()

        # Create the histogram plot with aggregated data
        plt.figure(figsize=(10, 6))
        sns.histplot(data=aggregated_data, x=metric, bins=50, kde=True)
    else:
        # Create the histogram plot without aggregation
        plt.figure(figsize=(10, 6))
        sns.histplot(data=combined_data, x=metric, bins=50, kde=True)

    plt.title(f'Distribution of {metric}')
    plt.xlabel(metric)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Define the engagement metrics
engagement_metrics = ['Engagements', 'Reactions', 'Likes', 'Comments', 'Shares', 'Post Clicks (All)']

# Group data by 'Post Type' and 'Network' and calculate the mean of engagement metrics
engagement_by_post_type = combined_data.groupby(['Content Type', 'Network'])[engagement_metrics].mean().reset_index()

# Plot engagement metrics over 'Post Type' and 'Network'
for metric in engagement_metrics:
    plt.figure(figsize=(12, 6))
    sns.barplot(data=engagement_by_post_type, x='Content Type', y=metric, hue='Network')
    plt.title(f'{metric} by Content Type and Network')
    plt.xlabel('Post Type')
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.legend(title='Network')
    plt.show()
```

```python
# Group data by 'Post Type' and calculate the mean of engagement metrics
engagement_by_post_type = combined_data.groupby('Post Type')[engagement_metrics].mean().reset_index()

# Sort by the most engaging post type for each metric
most_engaging_post_types = {}
for metric in engagement_metrics:
    most_engaging_post_types[metric] = engagement_by_post_type.sort_values(by=metric, ascending=False)['Post Type'].iloc[0]

# Display the most engaging post types for each metric
most_engaging_post_types
```

```python
# Group data by 'Network' and calculate the total engagement metrics
total_engagement_by_platform = combined_data.groupby('Network')[engagement_metrics].sum().reset_index()

# Sort by the platform with the highest total engagement for each metric
highest_engagement_platforms = {}
for metric in engagement_metrics:
    highest_engagement_platforms[metric] = total_engagement_by_platform.sort_values(by=metric, ascending=False)['Network'].iloc[0]

# Display the platform with the highest total engagement for each metric
highest_engagement_platforms
```

```python
# Plot the distribution of engagement by hour of the day
plt.figure(figsize=(12, 6))
sns.barplot(data=combined_data, x='Hour', y='Engagements', ci=None)
plt.title('Engagement by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Engagements')
plt.show()
```

```python
# Extract the year and month from the 'Date' column
combined_data['Year'] = combined_data['Date'].dt.year
combined_data['Month'] = combined_data['Date'].dt.month

# Group data by year and month and calculate the total Likes
likes_by_month = combined_data.groupby(['Year', 'Month'])['Likes'].sum().reset_index()

# Create a line plot of Likes over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=likes_by_month, x='Year', y='Likes', hue='Month', ci=None)
plt.title('Likes Over Time')
plt.xlabel('Year')
plt.ylabel('Total Likes')
plt.legend(title='Month')
plt.show()
```

12.	Content Type and Engagement by Network: LinkedIn excels in document and poll content engagement, while Facebook thrives in video, text, and other content engagement.

13.	Engagement by Hour: Engagement peaks during late-night and early morning hours.

14.	Trend Over Time: There is a notable increase in user reactions from 2020 to 2022.
